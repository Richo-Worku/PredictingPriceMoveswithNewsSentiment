```python
import os
import matplotlib.pyplot as plt

from scripts.technical_indicators import analyze_stock
```

```python
stocks = {}
import os

folder_path = "Data/"
for file in os.listdir(folder_path):

    if file.endswith(".csv"):

        stock_name = file.replace(".csv", "")

        file_path = os.path.join(folder_path, file)

        stocks[stock_name] = analyze_stock(file_path)
        for stock in stocks:
           stocks[stock] = stocks[stock].dropna()

print("All stocks processed successfully!")
import matplotlib.pyplot as plt

df = stocks["AAPL"]
df.isnull().sum()
df_clean = df.dropna()
plt.figure(figsize=(14,6))

plt.plot(df.index, df["Close"], label="Close Price")

plt.plot(df.index, df["sma_20"], label="SMA 20")

plt.plot(df.index, df["ema_20"], label="EMA 20")

plt.title("AAPL Moving Averages")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.show()
```
