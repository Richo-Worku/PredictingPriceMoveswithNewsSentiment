```python
stocks = {}
import os

folder_path = "Data/"
for file in os.listdir(folder_path):

    if file.endswith(".csv"):

        stock_name = file.replace(".csv", "")

        file_path = os.path.join(folder_path, file)

        stocks[stock_name] = analyze_stock(file_path)
     

print("All stocks processed successfully!")
import matplotlib.pyplot as plt

df = stocks["AAPL"]
df.isnull().sum()
df["sma_10"].head(20)
df_clean = df.dropna()
plt.figure(figsize=(14,6))

plt.plot(df.index, df["close"], label="Close Price")

plt.plot(df.index, df["sma_20"], label="SMA 20")

# plt.plot(df.index, df["ema_20"], label="EMA 20")
# plt.plot(df.index, df["macd"] , label="MACD")
# plt.plot(df.index,df["macd_signal"] , label="MACD SIGNAL")

plt.title("AAPL closing prices overlaid with moving averages")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()


##

plt.figure(figsize=(14,6))

plt.plot(df.index,  df["rsi"] , label="RSI")


# plt.plot(df.index, df["ema_20"], label="EMA 20")
# plt.plot(df.index, df["macd"] , label="MACD")
# plt.plot(df.index,df["macd_signal"] , label="MACD SIGNAL")



plt.title("AAPL RSI")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.show()

##
plt.figure(figsize=(14,6))

plt.plot(df.index,  df["macd"] , label="MACD")


# plt.plot(df.index, df["ema_20"], label="EMA 20")
# plt.plot(df.index, df["macd"] , label="MACD")
# plt.plot(df.index,df["macd_signal"] , label="MACD SIGNAL")



plt.title("AAPL MACD")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.show()



## PRICE RELATIONSHIP WITH INDICATORS

plt.figure(figsize=(14,6))

plt.plot(df.index, df["close"], label="Close Price")
plt.plot(df.index, df["sma_10"], label="SMA 10")
plt.plot(df.index, df["ema_10"], label="EMA 10")

plt.title("Price vs Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.xticks(rotation=45)


## PRICE RELATIONSHIP WITH RSI
fig, ax = plt.subplots(2, 1, figsize=(14,8))

ax[0].plot(df.index, df["close"], label="Close Price")
ax[0].set_title("Price")

ax[1].plot(df.index, df["rsi"], label="RSI", color="purple")
ax[1].axhline(70, linestyle="--")
ax[1].axhline(30, linestyle="--")
ax[1].set_title("RSI Indicator")

plt.tight_layout()
plt.show()

```
